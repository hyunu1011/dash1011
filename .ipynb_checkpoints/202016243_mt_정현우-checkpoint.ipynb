```python
df = pd.read_csv('https://raw.githubusercontent.com/guebin/DV2021/master/_notebooks/2021-10-25-FIFA22_official_data.csv').drop(['Loaned From','Marking'],axis=1).dropna()
df.head()
```

```python
position_dict = {
    'GOALKEEPER':{'GK'},
    'DEFENDER':{'CB','RCB','LCB','RB','LB','RWB','LWB'},
    'MIDFIELDER':{'CM','RCM','LCM','CDM','RDM','LDM','CAM','RAM','LAM','RM','LM'},
    'FORWARD':{'ST','CF','RF','LF','RW','LW','RS','LS'},
    'SUB':{'SUB'},
    'RES':{'RES'}
}
position_dict
```

```python
df2=df.set_axis(df.columns.str.replace(' ',''),axis=1)\
.loc[:,lambda _df: _df.isna().mean()<0.5].dropna()\
.assign(Position = lambda _df: _df.Position.str.split(">").str[-1].apply(lambda x: [k for k,v in position_dict.items() if x in v].pop()))\
.assign(Wage = lambda _df: _df.Wage.str[1:].str.replace('K','000').astype(int))\
.assign(Value = lambda _df: _df.Value.str[1:].str.replace('M','*1000000').str.replace('K','*1000').str.replace('€','').apply(eval).astype(int))\
.eval('logWage = log(Wage)').eval('logValue = log(Value)')
```

#### (4) 포워드와 수비수의 기술별 가치 평가 – 25점


```python
abilities_list = ['Crossing', 'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys', 'Dribbling', 'Curve', 'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration', 'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower', 'Jumping', 'Stamina', 'Strength', 'LongShots', 'Aggression', 'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure', 'StandingTackle', 'SlidingTackle']
abilities_categories = {
    "FinishingSkills": ['Finishing', 'HeadingAccuracy', 'Volleys', 'LongShots', 'Positioning', 'Vision', 'Penalties', 'ShotPower', 'Jumping'],
    "BallControl_Passing": ['Dribbling', 'Curve', 'Crossing','ShortPassing', 'LongPassing', 'BallControl','FKAccuracy'],
    "Speed_Stamina": ['Acceleration', 'SprintSpeed', 'Agility', 'Stamina'],
    "Reactions_PhysicalAttributes": ['Reactions', 'Balance', 'Strength', 'Composure'],
    "DefensiveSkills": ['Aggression', 'Interceptions', 'StandingTackle', 'SlidingTackle']
}
```

```python
def categorize_skills(row):
    for category, skills in abilities_categories.items():
        if row['SkillType'] in skills:
            return category
    return "Other"
```

```python
df3 = df2.melt(id_vars=['ID','Position','logValue'],value_vars= df2[abilities_list], var_name = 'SkillType'\
         ,value_name = 'SkillValue' ).assign(SkillTypeCategories = lambda df2 : df2.apply(categorize_skills, axis=1)
).sort_values(by='ID', ascending=True).assign(SkillValueCategories = lambda df2 : df2.groupby('ID')['SkillValue'].transform('mean'))
df3
```

```python
fig = ggplot(df3.query("Position=='FORWARD' or Position=='DEFENDER'"))
point = geom_point(aes(x='SkillValueCategories',y='logValue',color='Position'),alpha=0.05,size=0.05)
smooth = geom_smooth(aes(x='SkillValueCategories',y='logValue',color='Position'))
facet = facet_wrap('SkillTypeCategories')
fig = (fig + point + smooth + facet).draw()
fig.set_dpi(150)
fig.set_size_inches(8,5)
fig
```

## 3. 시각화의 해석 – 30점


#### (1) 심슨의 역설 – 10점

```python
df = pd.DataFrame({
    'Player': ['A', 'A', 'A', 'A', 'B', 'B', 'B', 'B'], 
    'Season': [1, 1, 2, 2, 1, 1, 2, 2],
    'Status': ['Success', 'Failure', 'Success', 'Failure'] * 2,
    'Count': [7, 3, 30, 1, 3, 1, 4, 0]
})
df
```

```python
df_A = df[df['Player'] == 'A']
df_B = df[df['Player'] == 'B']

# 시즌별 자유투 성공률 계산
seasonal_success_A = df_A[df_A['Status'] == 'Success'].groupby('Season')['Count'].sum()
seasonal_total_A = df_A.groupby('Season')['Count'].sum()
seasonal_free_throw_rate_A = seasonal_success_A / seasonal_total_A

seasonal_success_B = df_B[df_B['Status'] == 'Success'].groupby('Season')['Count'].sum()
seasonal_total_B = df_B.groupby('Season')['Count'].sum()
seasonal_free_throw_rate_B = seasonal_success_B / seasonal_total_B

# 전체 자유투 성공률 계산
total_success_A = df_A[df_A['Status'] == 'Success']['Count'].sum()
total_failure_A = df_A[df_A['Status'] == 'Failure']['Count'].sum()
total_free_throw_rate_A = total_success_A / (total_success_A + total_failure_A)

total_success_B = df_B[df_B['Status'] == 'Success']['Count'].sum()
total_failure_B = df_B[df_B['Status'] == 'Failure']['Count'].sum()
total_free_throw_rate_B = total_success_B / (total_success_B + total_failure_B)

# 데이터프레임 생성
data = pd.DataFrame({
    'Season': list(seasonal_free_throw_rate_A.index) + list(seasonal_free_throw_rate_B.index) + ['Total'] * 2,
    'Free Throw Rate': list(seasonal_free_throw_rate_A) + list(seasonal_free_throw_rate_B) + [total_free_throw_rate_A, total_free_throw_rate_B],
    'Player': ['A'] * len(seasonal_free_throw_rate_A) + ['B'] * len(seasonal_free_throw_rate_B) + ['A', 'B']
})

# 시각화
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='Season', y='Free Throw Rate', hue='Player')
plt.xlabel('Season')
plt.ylabel('Free Throw Rate')
plt.title('Seasonal and Total Free Throw Rates for A and B')
plt.show()
```
